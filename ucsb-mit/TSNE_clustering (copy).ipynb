# TSNE clustering algorithm

```python
# Omid55
# TSNE clustering code
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.manifold import TSNE
%matplotlib inline

tsne_model = TSNE(n_components=2)
#tsne_model = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)

def plot_TSNE(data, labels, names):
    points = 30 * tsne_model.fit_transform(data)
    df = pd.DataFrame(data=np.column_stack([points,labels]), columns=["x","y","class"])
    pl = sns.lmplot("x", "y", data=df, hue='class', fit_reg=False, palette=sns.color_palette('colorblind'), scatter_kws={"s": 70}, size=10, aspect=1)
    for i in range(len(points)):
        pl.ax.annotate(names[i], points[i])
    sns.plt.show()
    
def plot_tasks(fname):
    X = np.loadtxt(open(fname,"rb"),delimiter=",",skiprows=0)
    plot_TSNE(X, tlabels, tnames)
```
